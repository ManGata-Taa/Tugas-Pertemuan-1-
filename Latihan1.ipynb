5 nama mahasiswa

```python
mahasiswa = ["Andi", "Budi", "Citra", "Dewi", "Eko"]
print("List Mahasiswa: ")
print(mahasiswa)
```

Tambah dengan append

```python
mahasiswa.append("Fajar")
print("setelah ditambahkan nama baru:")
print(mahasiswa)
```

Hapus satu elemen menggunakan remove

```python
mahasiswa.remove("Citra")
print("Setelah menghapus nama Citra")
print(mahasiswa)
```
